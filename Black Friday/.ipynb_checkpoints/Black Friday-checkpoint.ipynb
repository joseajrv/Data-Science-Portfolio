# Black Friday Competition Project - Random Forest Regressor and GridSeach

A retail company “ABC Private Limited” wants to understand the customer purchase behaviour (specifically, purchase amount) against various products of different categories. 

They have shared purchase summary of various customers for selected high volume products from last month.

The data set also contains customer demographics (age, gender, marital status, city_type, stay_in_current_city), product details (product_id and product category) and Total purchase_amount from last month.

Now, they want to build a model to predict the purchase amount of customer against various products which will help them to create personalized offer for customers against different products.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('white')
```

Reading the train data:

```python
train = pd.read_csv('Data/Raw/train.csv')
```

```python
train.head(10)
```

The features of this dataset are:

Variable | Definition
--- | ---
User_ID	| User ID
Product_ID	| Product ID
Gender	| Sex of User
Age	| Age in bins
Occupation	| Occupation (Masked)
City_Category	| Category of the City (A,B,C)
Stay_In_Current_City_Years	| Number of years stay in current city
Marital_Status	| Marital Status
Product_Category_1	| Product Category (Masked)
Product_Category_2	| Product may belongs to other category also (Masked)
Product_Category_3	| Product may belongs to other category also (Masked)
Purchase | Purchase Amount (Target Variable)

Changes to be made:
Dummy variables or factorizing for Product_ID, Gender, City_Category and Ages.

Checking for missing data:

```python
null_values = pd.DataFrame(pd.isnull(train).any())
null_values['Sum'] = pd.isnull(train).sum()
null_values
```

Because the missing data is from categories, it may be important to keep it for the analysis, so the null values are going to be replaced by 0.

```python
train.info()
```

```python
train.fillna(0,inplace=True)
```

```python
train['Product_ID'] = pd.factorize(train['Product_ID'])[0]
```

```python
## Product_ID, Gender, City_Category and Ages
train['Age'] = pd.factorize(train['Age'])[0]
train['Gender'] = pd.factorize(train['Gender'])[0]
train['City_Category'] = pd.factorize(train['City_Category'])[0]
```

```python
train['Stay_In_Current_City_Years'] = pd.factorize(train['Stay_In_Current_City_Years'])[0]
```

```python
train.head(20)
```

```python
train.info()
```

Because the User_ID is a unique number is not related to the "Purchase" variable, it is discarded.

```python
train.drop('User_ID', axis=1,inplace=True)
```

Now cleaning the test data:

```python
test = pd.read_csv('Data/Raw/test.csv')
test.fillna(0,inplace=True)
dfp =  pd.read_csv('Data/Raw/test.csv')
test['Age'] = pd.factorize(test['Age'])[0]
test['Gender'] = pd.factorize(test['Gender'])[0]
test['City_Category'] = pd.factorize(test['City_Category'])[0]
test['Product_ID'] = pd.factorize(test['Product_ID'])[0]
test['Stay_In_Current_City_Years'] = pd.factorize(test['Stay_In_Current_City_Years'])[0]
```

```python
test.head()
```

```python
train.columns
```

Standardization of the variables:

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
scaled_train = scaler.fit_transform(train.drop('Purchase',axis=1))
scaled_train = pd.DataFrame(scaled_train,columns=['Product_ID', 'Gender', 'Age', 'Occupation', 'City_Category',
       'Stay_In_Current_City_Years', 'Marital_Status', 'Product_Category_1',
       'Product_Category_2', 'Product_Category_3'])
```

```python
train.head()
```

```python
scaled_train['Purchase'] = train['Purchase']
```

```python
scaled_train.head()
```

```python
X = scaled_train.drop('Purchase',axis=1)
y = scaled_train['Purchase']
```

This is a small version of the real param_grid which had a 'max_features': 'auto','log','sqrt', and number of estimators all the way to 100. 
Sadly it made the notebook extremely long with all the iterations so I had to cut it to the best parameters with just an example of another estimator to keep the grid search code:

```python
param_grid = {'n_estimators': [30,35],'max_features': ['auto']}
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
rfr = RandomForestRegressor(verbose = 2, oob_score = True, max_depth=10,n_jobs=4) 
```

```python
grfr= GridSearchCV(estimator=rfr, param_grid=param_grid)
```

```python
grfr.fit(X_train,y_train)
```

```python
predictions = grfr.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

As we can see, it is needed more feature engineering, because the beahavior tends to diverge

```python
grfr.best_params_
```

```python
grfr.best_score_
```

```python
grfr.best_estimator_
```

```python
from sklearn import metrics
print('MAE:',metrics.mean_absolute_error(y_test,predictions))
print('MSE:',metrics.mean_squared_error(y_test,predictions))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test,predictions)))
print('R^2:',metrics.explained_variance_score(y_test,predictions))
```

Now that we have trained the algorithm, got the features for the best estimation with this data, we know how it performs and satisfied with the precision, we can use it to get the predictions of the test data, which is the solution for the competition.

```python
X_train = scaled_train.drop('Purchase',axis=1)
y_train = scaled_train['Purchase']
test = test.drop('User_ID',axis=1)
```

```python
test.head()
```

```python
sns.heatmap(test.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
scaled_test = scaler.fit_transform(test)
scaled_test = pd.DataFrame(scaled_test,columns=['Product_ID', 'Gender', 'Age', 'Occupation', 'City_Category',
       'Stay_In_Current_City_Years', 'Marital_Status', 'Product_Category_1',
       'Product_Category_2', 'Product_Category_3'])
X_test = scaled_test
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=30,verbose = 2, oob_score = True, max_depth=10,n_jobs=4)
rfr.fit(X_train,y_train)
predictions = rfr.predict(X_test)
```

```python
df = pd.DataFrame(dfp['User_ID'])
df['Product_ID']=dfp['Product_ID']
df['Purchase']=predictions
df.to_csv("data/results/Submission Simple LR.csv", encoding='utf-8', index=False)
```
