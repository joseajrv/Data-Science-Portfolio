# Iris Dataset

The Iris dataset was used in R.A. Fisher's classic 1936 paper, The Use of Multiple Measurements in Taxonomic Problems, and can also be found on the UCI Machine Learning Repository.

It includes three iris species with 50 samples each as well as some properties about each flower. One flower species is linearly separable from the other two, but the other two are not linearly separable from each other.

The columns in this dataset are:

* Id
* SepalLengthCm
* SepalWidthCm
* PetalLengthCm
* PetalWidthCm
* Species

## Train Test Split

Now its time to split our data into a training set and a testing set!

** Using sklearn to split your data into a training set and a testing set as we've done in the past.**

```python
df_red.head()
```

```python
from sklearn.model_selection import train_test_split
X = df_red.drop('Species',axis=1)
y = df['Species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

## Training a K Means Clustering

** Import KMeans**

```python
from sklearn.cluster import KMeans
```

**Creating an instance of KMeans() and fitting it to the training data.**

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

Changing the labels 2 for 1 and the 1 for 2 in kmeans.labels_ to match the factorized labels of the dataframe.

```python
labels=[]
for value in kmeans.labels_:
    if value == 2:
        labels.append(1)
    elif value == 1:
        labels.append(2)
    else:
        labels.append(value)
print(labels)
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(X['PetalLengthCm'],X['PetalWidthCm'],c=labels,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(X['PetalLengthCm'],X['PetalWidthCm'],c=y,cmap='rainbow')
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y,labels))
```

96% using the petal related variables, now we are going to use all the variables to check the new results.

```python
X_2 = df.drop('Species',axis=1)
kmeans2 = KMeans(n_clusters=3)
kmeans2.fit(X_2)
kmeans2.cluster_centers_
kmeans2.labels_
```

```python
labels2=[]
for value in kmeans2.labels_:
    if value == 2:
        labels2.append(1)
    elif value == 1:
        labels2.append(2)
    else:
        labels2.append(value)
print(labels2)
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True,figsize=(10,6))
ax1.set_title('K Means 4 var')
ax1.scatter(X['PetalLengthCm'],X['PetalWidthCm'],c=labels2,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(X['PetalLengthCm'],X['PetalWidthCm'],c=y,cmap='rainbow')
ax3.set_title('K Means 2 var')
ax3.scatter(X['PetalLengthCm'],X['PetalWidthCm'],c=labels,cmap='rainbow')
```

```python
print(classification_report(y,labels2))
```

This time we got a worst precision. This is because, in this case, we already had 2 variables that almost defined(a precision of 96%),  which were the species of the flowers. 

So we could say that the most defining characteristic that separates each type of flower is the petal for the iris flowers. 

Having only these 2 variables for this particular dataset is beneficial given that the mere concept of K Means Clustering is putting the center of the clusters where the centroid is closest to each point of the group.
