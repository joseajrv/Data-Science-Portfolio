## The Data

We will be working with a famous titanic data set for these exercises. Later, we will revisit this data, and use it to predict survival rates of passengers. For now, we'll just focus on the visualization of the data with seaborn:

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
sns.set_context('paper')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.jointplot('fare','age',titanic)
```

```python
sns.distplot(titanic['fare'],kde=False,bins=30,color='red')
```

```python
sns.boxplot(x='class',y='age',data=titanic)
```

```python
sns.swarmplot(x='class',y='age',data=titanic)
```

```python
sns.countplot(x='sex',data=titanic)
```

```python
sns.heatmap(titanic.corr(),cmap='coolwarm')
plt.title('titanic.corr')
```

```python
g = sns.FacetGrid(titanic,col='sex')
g.map(plt.hist,'age')
```
